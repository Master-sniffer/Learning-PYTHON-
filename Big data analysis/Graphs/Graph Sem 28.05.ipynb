```python
import random
import networkx as nx
import matplotlib.pyplot as plt
```

# Part one

```python
class Chapter_one:
  def __init__(self):
    print("Hello there ! Here is the first part")
  def get_graph(self):
    return self.graph_one
  def add_nodes(self, graph_one):
    self.graph_one=graph_one
    self.graph_one.add_nodes_from(nodes_for_adding = list(range(100)))
    for i in range(150):
      if (i < 99):
        self.graph_one.add_edge(i, i+1)
      elif (i == 99):
        self.graph_one.add_edge(i, 0)
      else:
          first_side = random.choice(list(range(100)))
          second_side = random.choice(list(range(100)))
          while ((first_side, second_side) in list(self.graph_one.edges())) or ((second_side, first_side) in list(self.graph_one.edges())) or (first_side == second_side):
            first_side = random.choice(list(range(100)))
            second_side = random.choice(list(range(100)))
          self.graph_one.add_edge(first_side, second_side)
  def check_balance(self):
    balance = []
    for cycle in nx.cycle_basis(self.graph_one):
        stop = 1
        for node in range(len(cycle)-1):
            stop *= self.graph_one[cycle[node]][cycle[node + 1]]['sign']
        balance.append(stop)
    for num, cycle in enumerate(self.cyc):
        stop = 1
        for node in range(len(cycle)-1):
            stop *= self.graph_one[cycle[node]][cycle[node + 1]]['sign']
        balance.append(stop)
        str_sign = '+' if stop > 0 else '-'
        print(f'Cycle is {num + 1}, theeee sign is !!! - {str_sign}')
    if -1 in balance:
        print('Unbalanced')
    else:
        print('Balanced')
  def machinations(self):
    marks = [1, -1]
    G_edges = list(self.graph_one.edges())
    for edge in G_edges:
      self.graph_one[edge[0]][edge[1]]['sign'] = random.choice(list(marks))
  def draw_pic(self):
    positions = nx.spring_layout(self.graph_one)
    plt.figure(figsize = (12, 12))
    nx.draw_networkx(self.graph_one, positions)
    labels = nx.get_edge_attributes(self.graph_one, "sign")
    nx.draw_networkx_edge_labels(self.graph_one, pos=positions, edge_labels=labels)
    print(self.graph_one.number_of_nodes(), self.graph_one.number_of_edges())
  def additional_check(self):
    self.cyc = nx.cycle_basis(self.graph_one)
    print(len(list(self.cyc)))
    # for i in list(cyc): #not really matters
    #   print(i)
```

```python
c1 = Chapter_one()
graph_one = nx.Graph()
c1.add_nodes(graph_one)
c1.machinations()
c1.draw_pic()
c1.additional_check()
```

```python
c1.check_balance()
```
